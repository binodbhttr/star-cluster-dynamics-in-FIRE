```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
```

```python
path="./"
cluster_file_name="step586_3Myr_cp_m12i_res7100_mhdcv.cluster.v2.7100.b4.min4.pkl"
assoc_file_name="step586_3Myr_cp_m12i_res7100_mhdcv.cluster.v2.7100.b8.min4.pkl"
with open(path+cluster_file_name, "rb") as input:
    import_clusters = pickle.load(input)

print(import_clusters.keys())

with open(path+assoc_file_name, "rb") as input:
    import_assoc = pickle.load(input)

print(import_assoc.keys())
```

```python
print(import_assoc["vir_ratio"])
print(len(import_assoc["vir_ratio"]))
```

```python
fig1=plt.figure()
ax=fig1.add_subplot(1,1,1)
age_clusters=import_clusters["mean_age"]
y_clusters=age_clusters/import_clusters["r50"]
ax.scatter(age_clusters,y_clusters,s=3,c="red",alpha=0.2,label="Clusters")
ax.set_xlabel("Mean_Age")
ax.set_ylabel("Mean_Age/R50")
ax.set_title("Clusters And Associatons")
m_clusters, b_clusters = np.polyfit(age_clusters, y_clusters, 1)
sigma_clusters=np.std(y_clusters)
reg_y=m_clusters*age_clusters+b_clusters
ax.plot(age_clusters,reg_y,c="red",label="Fitted_Clusters")
ax.fill_between(age_clusters, reg_y - sigma_clusters, reg_y + sigma_clusters,
                 color='gray', alpha=0.2)

#overplotting the associations
age_assoc=import_assoc["mean_age"]
y_assoc=age_assoc/import_assoc["r50"]
ax.scatter(age_assoc,y_assoc,s=3,c="black",alpha=0.2,label="Associations")
m_assoc, b_assoc = np.polyfit(age_assoc, y_assoc, 1)
ax.plot(age_assoc, m_assoc*age_assoc+b_assoc,c="black",label="Fitted_Associations")
ax.set_ylim(0,2.0)
ax.legend(bbox_to_anchor=(1,0.5), loc='center left')




fig1.savefig("clusters_associations_analysis.jpeg",bbox_inches="tight",dpi=250)
```

```python
fig2=plt.figure()
ax=fig2.add_subplot(1,1,1)
age_assoc=import_assoc["mean_age"]
y_assoc=age_assoc/import_assoc["r50"]
ax.scatter(age_assoc,y_assoc,s=3)
ax.set_xlabel("Mean_Age")
ax.set_ylabel("Mean_Age/R50")
ax.set_title("Associations")

m, b = np.polyfit(age_assoc, y_assoc, 1)

#add linear regression line to scatterplot 
ax.plot(age_assoc, m*age_assoc+b,c="red")
```
